```python
import seaborn as sns
import pandas as pd

titanic_df = sns.load_dataset('titanic')
survivors_by_class = titanic_df.groupby(['class', 'survived']).size().reset_index(name='counts')
sns.barplot(x='class', y='counts', hue='survived', data=survivors_by_class)
```

```python
survivors_by_sex = titanic_df.groupby(['sex', 'survived']).size().reset_index(name='counts')
sns.barplot(x='sex', y='counts', hue='survived', data=survivors_by_sex)
```

```python
titanic_df['age_group'] = pd.cut(titanic_df['age'], bins=[0, 18, 60, 100], labels=['child', 'adult', 'elderly'])
survivors_by_age = titanic_df.groupby(['age_group', 'survived']).size().reset_index(name='counts')
sns.barplot(x='age_group', y='counts', hue='survived', data=survivors_by_age)
```

```python
titanic_df['family_size'] = titanic_df['sibsp'] + titanic_df['parch']
survivors_by_family_size = titanic_df.groupby(['family_size', 'survived']).size().reset_index(name='counts')
sns.barplot(x='family_size', y='counts', hue='survived', data=survivors_by_family_size)
```
