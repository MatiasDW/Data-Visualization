```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')
for column in iris.columns[:-1]:
    sns.displot(iris, x=column, kde=True)
    plt.show()
```

```python
for column in iris.columns[:-1]:
    sns.displot(iris, x=column, hue="species", kind="kde")
    plt.show()
```

```python
iris.groupby("species").agg(["mean", "median", "std"])
```

```python
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
plt.show()
```

```python
sns.scatterplot(data=iris, x="petal_length", y="petal_width", hue="species")
plt.show()
```

```python
new_iris = iris[(iris["petal_length"] > 5.5) & (iris["sepal_length"] > 4.5)]
sns.countplot(x="species", data=new_iris)
plt.show()
```

```python
sns.scatterplot(data=new_iris, x="petal_length", y="petal_width", hue="species")
plt.show()
```
